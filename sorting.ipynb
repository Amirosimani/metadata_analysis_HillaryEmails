```python
import csv
import pandas as pd
import numpy as np
import re
```

```python
#import the edge list
df = pd.read_csv("/Users/data_intern/Desktop/metadata_hillary-master/H_edgelist2.csv", encoding='utf-8')
pdf = pd.DataFrame(df)
```

```python
#cleaning up data frame by removing NaN, and number only entries
pdf = pdf.dropna()
pdf['from'] = pdf['from'].map(lambda x: re.sub('•', '', x))
pdf['to'] = pdf['to'].map(lambda x: re.sub('•', '', x))

pdf_sorted = pdf.sort_values(['from','to'], ascending= [1,1])
```

```python
#pdf_sorted = pdf_sorted.drop(pdf_sorted['from'].str.len() > 50)
#pdf_sorted = pdf_sorted.drop(pdf_sorted['to'].str.len() > 50)

pdf_sorted["to"] = pdf_sorted["to"].map(str.strip)
pdf_sorted["from"] = pdf_sorted["from"].map(str.strip)

pdf_sorted = pdf_sorted.dropna()
```

```python
pdf_sorted.to_csv('edge_list_python.csv', sep=',')
```
